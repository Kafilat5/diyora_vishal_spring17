Importing all needed module

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pylab, mlab
import random
import re
%matplotlib inline
```

```python
#Loading Housing Zhvi Price by state
census=pd.read_csv('Data/CSV/State_Zhvi_AllHomes.csv')
census.head()
```

```python
#Droping Unwanted Columns
columns_to_drop = list(range(1996,2008))
census= census[[col for col in census.columns if col[:4] not in str(columns_to_drop)]]
census.head()
# columns_to_drop
# census123=census
```

```python
gdp=pd.read_csv('Data/CSV/GDP_State_Quaterly.csv',skiprows=4,error_bad_lines=False)
#Removing punctation from columns
gdp.columns=gdp.columns.map(lambda x: re.sub(r'\W+', '', x))
#droping na
gdp.dropna(inplace=True)
columns_to_drop = list(range(2005,2008))
gdp= gdp[[col for col in gdp.columns if col[:4] not in str(columns_to_drop)]]



columnss=gdp.columns[2:]
columnss=columnss.tolist()
```

```python
#Cleaning and Manipulating the data
census2=census[census.columns.values[3:]]
census2.head()
```

```python
year=list(range(2008,2017))
census.columns.values[3:]
#Converting my data into quaters period
# census1=pd.to_datetime(census.columns.values[3:], format='%Y-%m').astype('period[Q]')
census1=census2.groupby(pd.PeriodIndex(census2.columns, freq='Q'), axis=1).mean()
#Here the manipulated data into Quaters.
census1.head()
```

```python
#Taking only 3 rows for the census dataframe
census=census[['RegionID','RegionName','SizeRank']]
census=census.join(census1)
census.head()
census.columns=census.columns.astype(str)
census.head()
```

```python
census_pivot=pd.melt(census,id_vars=['RegionName'],value_vars=columnss,var_name='Quater',value_name='Median_Price')
# census_pivot.head()
census_pivot=census_pivot.groupby(['RegionName','Quater'],).mean()
census_pivot.head()
```

```python
census_pivot=census_pivot.unstack(['RegionName'])
#Changing row and column data from the data base
census_pivot.head()
```

```python
gdp=pd.read_csv('Data/CSV/GDP_State_Quaterly.csv',skiprows=4,error_bad_lines=False)
#Removing punctation from columns
gdp.columns=gdp.columns.map(lambda x: re.sub(r'\W+', '', x))
#droping na
gdp.dropna(inplace=True)
columns_to_drop = list(range(2005,2008))
gdp= gdp[[col for col in gdp.columns if col[:4] not in str(columns_to_drop)]]

columnss
gdp.columns
```

```python
# columnss=columnss.tolist()
# # len(gdp.Area)
# columnss
#After Cleaning Data
gdp.head()
```

```python
gdp12=pd.melt(gdp,id_vars=['Area'],value_vars=columnss,var_name='Quater',value_name='Gdp')
#unstacking and transfering column and rows making gdp and housing dataframe row and column
gdp12.head()
```

```python
gdp12=gdp12.groupby(['Area','Quater'],).mean()
```

```python
# gdp12.index.astype(int)
```

```python
gdp12=gdp12.unstack(['Area'])
# gdp12.columns.names=['GDP','Area']
```

```python
gdp12.head()
```

```python
def gdpmedian(state):
    z=gdp12['Gdp'][state].plot(figsize=(10,4),color='r',label='GDP',title="GDP VS Median Price of {} ".format(state))
    # z.set_ylabel('GDP')
    z.set_xlabel('Month By Quater')
    z.set_ylabel('GDP {}'.format(state))
    # census_pivot['Median_Price']['California'].plot(figsize=(10,4),label='2017',title="GDP VS Median Price of California")
    # fig = plt.figure()
    # ax = fig.add_subplot(111)
    # z=a.plot(figsize=(10,4),label='2015',title="2008Mortgage 30 Year Fixed")

    # ax.plot(census_pivot.index,census_pivot['Median_Price']['California'],color='green' ,label = 'price')
    # x=mortgage1['MT']['2015'].plot(figsize=(8,4))
    # z.set_xlabel('Months')
    # z.set_ylabel('Median Price')
    # z.set_ylim(160000,180000)
    ax2=z.twinx()
    x=list(range(len(census_pivot.index)))
    ax2.plot(x,census_pivot['Median_Price'][state],color='g' ,label = 'MedianPrice')
    # ax2.xticks(x,census_pivot.index)
    # ax.set_xlim(1,)
    ax2.set_ylabel('Median House Price {}'.format(state))
    z.legend(loc=2)
    ax2.legend(loc=1)
    
```

```python

gdpmedian('New York')
# pylab.savefig('Analysis4/Analysis4_New_York.png')
```
